```python
import os
import xml.etree.ElementTree as xml
import lxml.etree as etree
```

```python
with open ("пример 11.txt") as f:
    filename = os.path.splitext(os.path.basename("пример 11.txt"))[0]
    root = xml.Element('{filename}'.format(filename=filename))
   # pages = xml.Element('Page')
   # root.append(pages)
    for line in f:
        line = line.rstrip()
        print(line)
        block = xml.SubElement(root, "block")
        block.text = line
        block.tail = '\n'

        
```

```python
tree = xml.ElementTree(root)
tree.write(open('from txt.xml', 'w'), encoding='unicode', xml_declaration=True, method="xml")
```
